```python
from Firebase import FirebaseChatManager

# Initialize the chat manager (replace with your actual credential path)
chat_manager = FirebaseChatManager(
    cred_path="nlp-assignment-c87d8-firebase-adminsdk-fbsvc-2eba3fbc80.json"
)

""" Testing firebase """
# Create a new chat session
user_id = "test_user_123"
chat_id = chat_manager.create_chat_session(user_id, "My First Chat")

print(f"Created chat with ID: {chat_id}")

# Add some messages
chat_manager.add_message(user_id, chat_id, "user", "Hello, how are you?")
chat_manager.add_message(user_id, chat_id, "bot", "I'm doing well, thanks!")
chat_manager.add_message(user_id, chat_id, "user", "What can you help me with today?")

# Retrieve and display messages
print("\nChat Messages:")
messages = chat_manager.get_chat_messages(user_id, chat_id)
for msg in messages:
    print(f"{msg['role']}: {msg['content']}")


# Get all chats for this user
print("\nUser's Chat Sessions:")
chats = chat_manager.get_user_chats(user_id)
for chat in chats:
    print(f"Chat ID: {chat['id']}, Name: '{chat['chat_name']}', Messages: {chat['message_count']}")

# Get specific chat details
print("\nChat Details:")
chat_details = chat_manager.get_chat_details(user_id, chat_id)
print(chat_details)


#------------------------------------------------------------------
# Get all chats for this user
print("\nAll Chat Sessions with Messages:")
chats = chat_manager.get_user_chats(user_id)

for chat in chats:
    print(f"\nChat ID: {chat['id']}")
    print(f"Chat Name: '{chat['chat_name']}'")
    print(f"Total Messages: {chat['message_count']}")
    print("Messages:")
    
    # Get messages for this specific chat
    messages = chat_manager.get_chat_messages(user_id, chat['id'])
    
    if not messages:
        print("  No messages in this chat")
    else:
        for msg in messages:
            # Handle both 'role' and 'sender' fields for backward compatibility
            role = msg.get('role', msg.get('sender', 'unknown'))
            print(f"  {role}: {msg.get('content', '')}")
    
    print("-" * 40)  # Separator between chats
```

```python
# Clean up (uncomment when needed)
chat_manager.delete_chat(user_id, chat_id)
print("Chat deleted")
```
